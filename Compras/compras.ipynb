```python
import psycopg2
import random
import time
from faker import Faker

fake = Faker('pt_BR')
```

```python
def gerador_de_dados_em_massa(quantidade_total):
    """
    Função que gera e salva uma quantidade predefinida de compras.
    """
    print(f"🚀 Iniciando a gravação de {quantidade_total} registros...")
    
    conexao = None
    
    try:
        # Abrimos a conexão UMA VEZ antes de começar o loop
        conexao = psycopg2.connect(
            user="postgres",
            password="0542",
            host="localhost",
            port="5432",
            database="Data_lake_supermercado",
            options="-c lc_messages=C"
        )
        cursor = conexao.cursor()
        
        contador = 0
        
        # O LOOP CONTROLADO
        while contador < quantidade_total:
            # 1. Gerando dados aleatórios
            codigo_barra = fake.ean13()
            nome_produto = fake.word().capitalize()
            tipo_produto = random.choice(['Limpeza', 'Alimento', 'Bebida', 'Higiene', 'Padaria'])
            preco_unitario = round(random.uniform(1.99, 89.90), 2)
            nome_caixa = fake.name()
            numero_caixa = random.randint(1, 10)
            data_da_compra = fake.date_between(start_date='-1m', end_date='today')

            # 2. Preparando o SQL
            comando_sql = """
                INSERT INTO compras (
                    codigo_barra, nome_produto, tipo_produto, 
                    preco_unitario, nome_caixa, numero_caixa, data_da_compra
                ) VALUES (%s, %s, %s, %s, %s, %s, %s);
            """
            valores = (codigo_barra, nome_produto, tipo_produto, preco_unitario, nome_caixa, numero_caixa, data_da_compra)
            
            # 3. Executando e Salvando
            cursor.execute(comando_sql, valores)
            conexao.commit() # Salva cada registro individualmente
            
            contador += 1 # Aumenta o contador para não ter loop infinito
            
            print(f"[{contador}/{quantidade_total}] ✅ Salvo: {nome_produto} | Caixa {numero_caixa}")
            
            # 4. A PAUSA DE 1 SEGUNDO
            if contador < quantidade_total: # Se ainda não for o último, espera
                time.sleep(0.0001)

        print(f"\n--- ✨ Sucesso! {quantidade_total} registros foram inseridos. ---")

    except (Exception, psycopg2.Error) as error:
        print("❌ Erro durante a execução:", error)
        if conexao:
            conexao.rollback()
        
    finally:
        # Fechando a conexão apenas no final de tudo
        if conexao:
            cursor.close()
            conexao.close()
            print("🔌 Conexão com o PostgreSQL encerrada com segurança.")

# --- EXECUÇÃO ---
# Altere o número abaixo para quantos registros você deseja gerar
gerador_de_dados_em_massa(100)
```
